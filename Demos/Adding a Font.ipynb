```python
import matplotlib.pyplot as plt
from matplotlib import font_manager
```

```python
# https://fonts.google.com/specimen/Grenze+Gotisch?category=Display
locations = ['Downloads/Grenze_Gotisch']
```

```python
font_files = font_manager.findSystemFonts(fontpaths = locations)
```

```python
for file in font_files:
    font_manager.fontManager.addfont(file)
```

```python
# example

fig, ax = plt.subplots(figsize = (12,3), facecolor = (.2, .1, .1))

ax.axis('off')
fig.text(0.5, 0.5, 'Matplotlib for Storytellers', ha = 'center', fontname = 'Grenze Gotisch', 
        size = 50, fontweight = 'extra bold', color = 'white', va = 'center')
```

```python
# example

fig, ax = plt.figure(figsize = (12,8)), plt.axes(facecolor = (.2, .1, .1))

ax.set_xticks([])
ax.set_yticks([])

for key in 'top', 'bottom', 'left', 'right':
    ax.spines[key].set_linewidth(10)
    ax.spines[key].set_color("gold")
    
ax.text(0.5, 0.5, 'Matplotlib\nfor\nStorytellers', 
            ha = 'center', va = 'center',
            fontname = 'Grenze Gotisch', 
            size = 68, fontweight = 'extra bold', 
            color = 'white')    

plt.show()
```
